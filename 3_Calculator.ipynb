# 간단한 사칙연산
##사칙연산

- 정수와 실수
- 정수 : 자연수와 자연수의 음수, 그리고 0을 이루어진 수의 집합
- 실수 : 유리수와 무리수의 이루어진 집합
- 프로그램에서는 숫자에 소수점이 없으면 정수(int)이고 소수점이 있으면 실수(flot)
- 실수 연산만 할 수 있는것이 아니라 실수 연산도 할 수 있음
- 연산 기호가 두 개 이상일 경우 일반적인 연산 규칙

```python
print(1.2+5.3)
print(2*2)
print(10/2)
print((5*4-15)+((5-2)*(9-7)))
```

## 과학적 표기법

- 연산할 때 아주 큰 수나 작은 수를 다뤄야 할 때 특히 과학이나 공학 분야
- 거듭 제곱 연산자로 큰 수를 입력할 수 있지만 10의 거듭제곰(즉, 10n)의 경우 `e`n으로 편리하게 입력, 여기서 n은 정수로 양수, 음수, 0
- 10의 거듭제곱 표시를 위한 `e`n 앞에는 항상 숫자가 있어야 함
- 과학적 표기법으로 출력할 때는 e 다음에는 지수가 양수냐 음수냐에 따라서 양수 기호(+)나 음수 기호(-)가 들어감

```python
print(3*10**8)
print(3e8)
print(1e-3)
print(1e-4)
print(1e-5)
print(1e-6)
```

## 진수 표현과 변환

- 2진수, 8진수, 16진수 형태로 숫자를 입력해도 기본적으로 10진수로 출력
- 10진수 외에` 2진수(0,1)`, `8진수(0~7)`,`16진수(0~9,a~f)`를 입력하기 위해서는 숫자 앞에 각각 `0b`, `0o`,`0x` 를 붙임
- 2진수, 8진수, 16진수로 변환하는 함수는 각각의` bin()`, `oct()`, `hex()`
- 출력 결과는 숫자가아니라 문자 열
- 작은 따움표('') 안에 있음
- int('수',진법) -> 10진수로 변경

```python
print(17)
print(bin(17))
print(oct(17))
print(hex(17))
print(int('0b10001',2)) #int('수',진법) -> 10진수로 변경
```

## 논리연산
- 어떤 조건을 참과 거짓을 이용해 연산하는 논리 연산(logical operation), 논리 연산은 불린 연산(Boolean operation)이라고도 함
- 논리 연산을 위한 데이터 타입은 불(bool)
- 불 테이터 타입에는 논리 참 혹은 논리 거짓이 있음
- 따옴표를 이용해 입력하면 불 데이터가 아니라 문자열 데이터로 인식함
- 불 데이터의 경우 논리 연산만 할 수 있음
캡처.PNG

```python
print(type(True)) #bool
print(True and False)
print(True and True)
print(True or False)
print(not True)
```

## 비교 연산자
- 비교 연산자의 결과는 불 데이터 형식으로 나옴

```python
print(5==3)
print(5!=3)
print(5<3)
print(5>3)
print(5<=3)
print(5>=3)
```

## 논리 연산과 비교연산자 혼합연산
- 비교 연산자와 논리 연산자가 함께 있으면 비교 연산을 먼저 함
- 괄호가 여러 겹 있을 때는 가장 안쪽 괄호부터 먼저 계산함

```python
print(1>0 and -2<0)
print(1>0 or -2<0)
print((3>0) or ((-5>0) and (1>5)))
print(((3>0) or (-5>0)) and ((4>8) or (3<0)))
```
