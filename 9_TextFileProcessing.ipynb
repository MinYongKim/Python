##9.1 split()
- 구조식 - str.split(sep)
- 문자열(str)에서 구분자(sep) 기준으로 문자열 분리
- str.split() 수행시 문자열 사이의 모든 공백과 개행문자(\n)을 삭제 분리된 문자열을 항목으로 리스트로 반환

- str.split(sep, maxsplit=숫자
    - 앞에서부터 원하는 횟수 만큼만 문자열을 분리 할 수 있음

```python
# str,split()
Nike = "I can do it"
Nike.split()
```

```python
# str,split(sep,maxsplit)

p_num = "+82-01-1234-5678"
split_num = p_num.split("-", 1)
print(split_num)
print("국내전화번호 : {0}".format(split_num[0])) # 리스트 반환으로 인덱스 출력가능
```

##9.2 strip()
- 구조식 - str.strip()
- 필요없는 문자열(공백, \n) 앞과 뒤 삭제
- 모두 지정해서 한번에 제거 가능
- 문자열 왼쪽(앞) lstrip()
- 문자열 오른쪽(뒤) rstrip()

```python
text_str = "aaaabbbbbbbcccccccPythoncccccbbbbbbaa"
text_str1 = text_str.strip('ab')
print(text_str1)
text_str2 = text_str1.strip('c')
print(text_str2)
```

```python
text_str = "abcabcPythoncbaabc"
text_str1 = text_str.lstrip('abc')
print(text_str1)
text_str2 = text_str1.rstrip('abc')
print(text_str2)
```

##9.4 find() / count() / startwith() / endwith()
### str.find(search_str)
- 문자열의 위치는 0 부터 시작
- 찾을 수 없으면 -1 을 반환
- 특정 문자열을 찾아서 맨 처음 발견된 위치
- 범위 설정
    - str(search_str, start, end)
    - 범위에서 일치하는 문자열 위치 반환
    
### str.count(search_str)
- 문자열이 몇 번 나오는지 확인

### str.startwith() / str.endwith()
- 문자열이 지정된 문자로 시작/끝이 나는지 검사
- True / Flase 반환

```python
ser_Str = "Python is powerful. Python is easy to learn"

print(ser_Str.find("Python", 10, 30))
print(ser_Str.find("were"))
print("Python의 문자의 개수:? ", ser_Str.count("Python"))
print("learn의 문자의 개수:?", ser_Str.count("learn"))
print(ser_Str.startswith("Python"))
print(ser_Str.endswith("learn"))
```

## 9.6 데이터 파일 다루기
- 파일의 내용을 한 번에 읽어 오는것이 아니라 한 줄씩 읽어서 처리
- 데이터를 원하는 형태로 전처리


```python
!cat coffeeShopSales.txt
```

```python
file_name = 'coffeeShopSales.txt'
```

```python
f = open(file_name)
for line in f :
    print(line, end = '')
f.close()
```

```python
f = open(file_name)
header = f.readline()
header_list = header.split()

espresso=[]
americano=[]
cafelatte=[]
cappucino=[]
for line in f:
    data_list = line.split()
    espresso.append(int(data_list[1]))
    americano.append(int(data_list[2]))
    cafelatte.append(int(data_list[3]))
    cappucino.append(int(data_list[4]))

f.close()

print("{0}: {1}".format(header_list[1], espresso))
print("{0}: {1}".format(header_list[2], americano))
print("{0}: {1}".format(header_list[3], cafelatte))
print("{0}: {1}".format(header_list[4], cappucino))
    
```

```python
total_sum = [sum(espresso), sum(americano), sum(cafelatte), sum(cappucino)]
total_mean = [sum(espresso)/len(espresso), sum(americano)/len(americano),
              sum(cafelatte)/len(cafelatte), sum(cappucino)/len(cappucino)]

for k in range(len(total_sum)):
    print('[{0}] 판매량'.format(header_list[k+1]))
    print('- 4일 전체 : {0}, 하루평균 : {1} '.format(total_sum[k], total_mean[k]))
```
